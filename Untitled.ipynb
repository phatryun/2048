```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
print("score Alea")
df_alea = pd.read_csv("./result_naif_IA.csv").reset_index()
df_alea.head()
```

```python
df_alea_tile = df_alea.groupby(["highest_tile"], as_index=False).agg({"score":"count", "nb_moves":("sum", "mean")})
df_alea_tile
```

```python
fig, ax = plt.subplots()

df_alea["highest_tile"].plot(ax=ax, style='b-')
df_alea["nb_moves"].plot(ax=ax, style='r-',  secondary_y=True)

plt.show()
```
